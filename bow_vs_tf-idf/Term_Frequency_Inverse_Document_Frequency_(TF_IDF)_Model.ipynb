```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Data review film
data = [
    "The movie was fantastic",
    "I loved the storyline but the acting was bad",
    "I didn't like the movie at all"
]

# Menampilkan data review dalam bentuk DataFrame
df_reviews = pd.DataFrame(data, columns=['Reviews'])
display(df_reviews)

# Membuat objek TfidfVectorizer
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(data)

# Menampilkan nama fitur dalam bentuk DataFrame
df_features = pd.DataFrame(vectorizer.get_feature_names_out(), columns=['Feature Names'])
display(df_features)

# Menampilkan matriks TF-IDF dalam bentuk DataFrame
df_tfidf = pd.DataFrame(tfidf.toarray(), columns=vectorizer.get_feature_names_out())
display(df_tfidf)
```
