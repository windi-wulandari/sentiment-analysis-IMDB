```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# List of movie reviews
reviews = [
    "The movie was fantastic",
    "I loved the storyline but the acting was bad",
    "I didn't like the movie at all"
]

# Converting reviews into a DataFrame
reviews_df = pd.DataFrame(reviews, columns=["Reviews"])

# Print and display the reviews DataFrame with a line break
print("\nReviews DataFrame:")
display(reviews_df)

# Initialize CountVectorizer to transform text data into a vector of word counts
vectorizer = CountVectorizer()

# Fit and transform the reviews into word count vectors
X = vectorizer.fit_transform(reviews)

# Convert the transformed data into a DataFrame with words as columns
df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Print and display the DataFrame showing word counts for each review with a line break
print("\nWord Count DataFrame:")
display(df)
```
